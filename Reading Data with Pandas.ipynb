# 1. Reading diffrent type of file

Methods to read data are all named **pd.read_***  where * is the file type. for Example 

* Text Files: **read_csv**, **read_fwf** (Fixed-Width Text File),**read_json**,**read_html**,**read_clipboard** 
* Excel File: **read_excel**
* Other file formats: **read_hdf** (HDF5),**read_feather** (Feather), **read_parquet** (Parquet) ,**read_orc** (ORC),**read_msgpack** (Msgpack),**read_stata** (Stata),**read_sas** (SAS),**read_spss** (SPSS) ,**read_pickle** (Python Pickle), **read_sql** (SQL), **read_gbq** (Google BigQuery)

## 1.1 Reading Text File formats

```python
import pandas as pd
```

```python
df_csv = pd.read_csv('Titanic.csv', sep=',',  parse_dates=True, )
"""
These are the important arguments:
• filepath_or_buffer: Either a path to a file , URL (including http, ftp, and S3 locations), 
or any object with a read() method (such as an open file or StringIO).
• sep : seprater(delimiter)
• names: set or override column names
• parse_dates: accepts multiple argument types, see on the right
• converters: manually process each element in a column
• comment: character indicating commented line
"""
```

```python
df_csv.head()
```

```python
widths = [10,5,5,5,6,5,7,7,4,5]
df_fwf  = pd.read_fwf('test_fwf.csv', widths=widths, header=None)
"""
 As per your need you  can create a widths List.
 here I have created a sample test file in above example.
 you can check the repository and see the 'test_fwf.csv' file 
"""
```

```python
df_fwf.head()
```

```python
df_html = pd.read_html('https://en.wikipedia.org/wiki/Coronavirus_disease_2019', match='Symptoms of *')
"""
These are the important arguments:
•match : str or compiled regular expression
•attrs : This is a dictionary of attributes that you can pass to use to identify the table in the HTML. 
"""
```

```python
df_html[1]
```

```python
df_json =  pd.read_json('Titanic.json', orient="columns")
"""
orient: Indication of expected JSON string format.
The set of possible orients is:
'split' : dict like {index -> [index], columns -> [columns], data -> [values]}
'records' : list like [{column -> value}, ... , {column -> value}]
'index' : dict like {index -> {column -> value}}
'columns' : dict like {column -> {index -> value}}
'values' : just the values array

convert_dates:bool or list of str (convert_dates can convert various format of dates)
"""
```

```python
df_json.head()
```
